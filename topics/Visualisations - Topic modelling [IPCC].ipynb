### Creating topic name dataframe

```python
combined_topics_final_reps = combined_topics_final_reps.transpose()
combined_topics_final_reps = combined_topics_final_reps.iloc[2: , :]
combined_topics_final_reps = combined_topics_final_reps.reset_index()
combined_topics_final_reps = combined_topics_final_reps.rename(columns={'index': 'Topic'}) 
# combined_topics_final_reps
```

```python
by_topic_count = combined_topics_final.groupby(['Topic','Name']).agg(topic_count = ('text','count'))
by_topic_count['%_total'] = round(by_topic_count['topic_count'] / by_topic_count['topic_count'].sum()*100,0)
by_topic_count = by_topic_count.reset_index()
# by_topic_count
```

```python
combined_topics_final_reps['Topic'] = combined_topics_final_reps['Topic'].astype(int) 
by_topic_count['Topic'] = by_topic_count['Topic'].astype(int) 

df_topics = combined_topics_final_reps.merge(by_topic_count, on = 'Topic',how='inner')

df_topics = df_topics.rename(columns={'0': 'Document1','1': 'Document2','2': 'Document3','Name': 'Top_words'})
#df_topics
df_topics.to_csv('ipcc_topics_final_reps_fixed2.csv')
```

```python
df_topics['Top_words'] = df_topics['Top_words'].str.replace('_',', ').str.split(',', n=1).str.get(-1)
```
